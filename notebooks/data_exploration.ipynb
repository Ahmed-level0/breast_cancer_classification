```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
data = pd.read_csv('../data/data.csv')  
```

```python
sns.set_style('darkgrid')
```

```python
data.head().T
```

>dropping the weird column

```python
data['diagnosis'] = data['diagnosis'].map({'B': 0, 'M': 1})
```

```python
data.info()
```

```python
data.drop(columns=['Unnamed: 32', 'id'], inplace=True)
```

```python
data.describe().T
```

```python
num_cols = data.drop(columns=['diagnosis', 'id']).columns
num_cols
```

```python
n = len(num_cols)
rows = int(np.ceil(n/4))

fig, axs = plt.subplots(rows, 4, figsize=(20, 5*rows))

# Flatten the axes array, to make iterating over it easier
axs = axs.flatten()

for i, c in enumerate(num_cols):
    sns.histplot(x=data[c], ax=axs[i], kde= True, hue=data['diagnosis'], palette='Set2')
    axs[i].set_title(f'distribution of {c}')

# Remove any unused subplots
if n % 4:
    for j in range(n, rows*4):
        fig.delaxes(axs[j])

plt.tight_layout()
plt.suptitle('Distribution of Numerical columns', x=0.5, y=1.02, fontsize=16)
plt.show()
```

```python
skewed_cols = []
for col in num_cols:
    if data[col].skew() > 1:
        skewed_cols.append(col)
```

```python
sns.countplot(data= data, x= 'diagnosis', palette='Set2', hue= 'diagnosis')
plt.title("Diagnosis (1) for Malignant and (0) for Benign")
plt.show()
```

```python
data.corr(numeric_only=True)
```

```python
sns.heatmap(data.corr(numeric_only=True), cmap='flare')
plt.title("Correlation Heatmap")
plt.show()
```

```python
for col in num_cols:
    for col2 in num_cols:
        print(col, data[col].corr(data[col2]), col2)
```
